```python
import pandas as pd
import glob
import os

path = r'C:\Users\Admin\Downloads\Banknifty_data' # use your path
all_files = glob.glob(os.path.join(path, "*.csv"))
# print(all_files)
for filename in all_files :
        data = pd.read_csv(filename, index_col=None, header=0)
        future_data = data[(data['instrument']=='FUTIDX') & (data['option_type']=="XX")] 
        future_data.reset_index(drop=True,inplace=True)
        Future_price = future_data[future_data['time']=='09:19:00'].iloc[0,11]
        Future_expiry = future_data[future_data['time']=='09:19:00'].iloc[0,5]
        base_price = 100

        a = round(Future_price/base_price)
        atm = (a*base_price)
        atm
        
        ce_data = data[(data['option_type']=='CE') & (data['strike_price']==atm) & (data['exp_date']==Future_expiry)] 
        ce_data.reset_index(drop=True,inplace=True)
        ce_data = ce_data.sort_values(by = ['time'] , ascending = True)
        pe_data = data[(data['option_type']=='PE') & (data['strike_price']==atm) & (data['exp_date']==Future_expiry)] 
        pe_data.reset_index(drop=True,inplace=True)  
        pe_data = pe_data.sort_values(by =['time'],ascending = True)
        for i in range(len(ce_data)):
            if ce_data['time'][i]=='09:19:00' :
                print(ce_data['date'][i], ce_data['close'][i]+pe_data['close'][i])
```
